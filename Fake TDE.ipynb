## Plotting a Fake TDE to Use for Machine Learning

```python
#plotting a fake TDE
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import statistics
from statistics import median
import time
```

```python
plt.figure(figsize = (10,10))

time = [-6, -3, 0, 1, 4, 5, 6, 7, 8, 9, 10, 24, 26, 30, 34, 38, 42, 48, 50, 56, 62, 64, 66]

mag = [17.4, 17.1, 17, 17.1, 17, 17.2, 17.25, 17.3, 17.35, 17.4, 17.45, 18.6, 18.8, 19.4, 18.6, 19.2, 19.3, 19.2, 24, 19.6, 20, 19.9, 19.4]

plt.gca().invert_yaxis()
plt.xlabel('Time (MJD - 57632.1)')
plt.ylabel('Magnitude')
plt.title('Plot of Time vs. Magnitude for Defined TDE')
plt.scatter(time, mag, color = 'green')
```

```python
plt.figure(figsize = (10,10))

time = [-6, 0, 5, 24, 56, 62]

mag = [17.4, 17, 17.2, 19.2, 19.6, 20]

mag_err = [0.03, 0.05, 0.001, 0.001, 0.2, 0.1]

time_mag_err = tuple(zip(time, mag, mag_err))

plt.gca().invert_yaxis()
plt.xlabel('Time (MJD - 57632.1)')
plt.ylabel('Magnitude')
plt.title('Plot of Time vs. Magnitude for Defined TDE')

for i, value in enumerate(time_mag_err):
    
    plt.scatter(time_mag_err[i][0], time_mag_err[i][1])
    plt.errorbar(time_mag_err[i][0], time_mag_err[i][1], yerr = time_mag_err[i][2], fmt = 'o', color = 'green')
    
plt.show()
```
